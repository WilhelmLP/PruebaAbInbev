```python
import pandas as pd


def transform_parquet_to_csv(file: pd.DataFrame) -> pd.DataFrame:
    """Transforma un archivo .parquet a .csv

    Args: 
        file: archivo.parquet

    Return: Recibe como parametro un archivo parquet de tipo DataFrame para devolver un DataFrame

    """

    df = pd.read_parquet(file, engine="pyarrow")
    return df.to_csv('dataset.csv', index=False)


def divice_to_csv(file: pd.DataFrame, cat: str) -> pd.DataFrame:
    """Divide a .csv by category and return files separated by .csv 

    Args: 
        file: archivo.csv
        cat: name of category for select the separete
    
    Return:  file in format .csv
    
    """

    for i in file.Category.unique():
        file_2= file[file[f"{cat}"] == i]
        file_2.to_csv(f'{i}.csv', index=False)
    return file_2
```
