```python
"""
In this jupyter notebook the exploratory analysis of the data is done
to be able to identify its particular dataset and its correlations
""" 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
plt.rcParams['lines.linewidth'] = 1.5
```

```python
filename = '../inputs/Lacteos.csv'
df = pd.read_csv(filename)
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
print(f'Missing values: {df.isnull().any(axis=1).mean()}')
```

```python
pd.set_option('display.width',
             100)
df.describe()
```

```python
df.groupby('Category').size()
```

```python
pd.set_option('display.width',
               100)
correlation = df.corr(method='pearson')
print(correlation)
```

```python
%matplotlib inline
correlation = df.corr()
plt.figure(figsize=(7,7))
ax = sns.heatmap(correlation, vmax=1, square=True, annot=True, cmap="viridis")
plt.title("Correlation")
```

```python
fig = plt.figure(figsize = (7,7))
ax = fig.gca()
df.hist(ax = ax)
```
